```python
import pandas as pd
```

```python
props = pd.read_csv('http://www.firstpythonnotebook.org/_static/committees.csv')
contribs = pd.read_csv('http://www.firstpythonnotebook.org/_static/contributions.csv')
```

```python
# prints out first five rows of the dataframe
# defaults to five, accepts input of how many rows
# ex. df.head(7) will print seven rows
props.head(3)
```

```python
# .info() gives a snapshot of the data
# rangeindex gives number of rows
# prints columns, num of inputs and associated data types
props.info()
```

```python
# .value_counts returns a series
# gives the value and count of apperances
props.prop_name.value_counts()
```

```python
# Pandas uses indexing to filter data
prop64 = props[props.prop_name == 'PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE.']

# Best practice (http://pandas.pydata.org/pandas-docs/stable/indexing.html)
prop64 = props[props['prop_name'] == 'PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE.']  


```

```python
prop64.info()
```

```python
contribs.info()
```

```python
# join types
# inner join: only records where the identifyers match are kept. could result in fewer records in the join system than the two separate.
# outer join: all identifyers are kept regardless of being matched in both

# merge
merged = pd.merge(prop64, contribs, on='calaccess_committee_id', how='inner')
```

```python
merged.head()
```

```python
merged.info()
```

```python
# .sum()
# good to note that sum won't *always* be accurate depending on the data 
# so be careful with framing. this isn't the complete sum because donors 
# who contributed less than $100 were left off the contribution list.
merged.amount.sum()
```

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged['committee_position'] == 'SUPPORT']
oppose = merged[merged.committee_position == 'OPPOSE']
```

```python
support['amount'].sum()
```

```python
oppose.amount.sum()
```

```python
support['amount'].sum() / merged.amount.sum()
```

## Sorting

```python
merged.sort_values('amount', ascending=False)
```

```python
oppose.sort_values('amount', ascending=False)
```

## Groupby

```python
# groupby will not display anything by itself
# must be chained with a calculation
merged.groupby('committee_name_x').amount.sum().reset_index().sort_values('amount', ascending=False)
```

```python
merged.groupby([ 'contributor_firstname', 'contributor_lastname', 'committee_position' ]).amount.sum().reset_index().sort_values('amount', ascending=False)
```
